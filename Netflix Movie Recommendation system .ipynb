```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Assuming you've already calculated your vectors and similarity matrix as in your previous code
# cv = CountVectorizer(max_features=5000, stop_words='english')
# vectors = cv.fit_transform(new_df['tags']).toarray()
# similarity = cosine_similarity(vectors)

def recommend_movies(movie_title, vectors, similarity_matrix, top_n=5):
    # Find the index of the input movie title
    movie_index = new_df[new_df['title'] == movie_title].index[0]

    # Calculate similarity scores for the input movie
    similarity_scores = similarity_matrix[movie_index]

    # Enumerate the similarity scores and sort them by value in descending order
    sorted_indices = sorted(enumerate(similarity_scores), key=lambda x: x[1], reverse=True)

    # Extract the top N indices from the sorted list (excluding the input movie itself)
    top_indices = [index for index, _ in sorted_indices if index != movie_index][:top_n]

    # Get the titles of the recommended movies
    recommended_movies = new_df['title'].iloc[top_indices]

    return recommended_movies



```
